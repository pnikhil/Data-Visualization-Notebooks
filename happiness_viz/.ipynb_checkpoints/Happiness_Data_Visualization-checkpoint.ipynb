# HAPPINESS DATA VISUALIZATION

The world happines report dataset was first published in 2012. It is a survey released by the United Nations. It ranks 155 countries by their happiness using data from the Gallup World Poll.

The happiness scores are based on answers to main life evaluation question.This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale.

The other columns are the following factors per country:
-Economic production
-Social support(Family)
-Life expectancy
-Freedom
-Absence of corruption
-Generosity

These six factors have no impact in the total score reported for each country. Each factor have positive values visualizing how higher a country is in terms of each specific factor than Dystopia.

Dystopia is an imaginary country with the least happy people which factors are equal to those of the country with less happiness in the list.

Source: https://www.kaggle.com/unsdsn/world-happiness

Converted data to a tidy data set by removing all the redundant rows or columns that weren't appropriate. Also made sure that the columns were renamed to something more apt for  visualizations. The data is now organised and ordered. All three data files were compared and made tidy such that there are no repetitions or redunant data.

```python
#loading the tidy data
dataframe_2015 = pd.read_csv('data_sources/2015_cleaned.csv')
dataframe_2016 = pd.read_csv('data_sources/2016_cleaned.csv')
dataframe_2017 = pd.read_csv('data_sources/2017_cleaned.csv')
```

```python
#summary of data

dataframe_2017.info()
```

```python
#Changing column names for easier processing

dataframe_2015.columns = ("Country", "Region", "Happiness_Rank_15", "Happiness_Score_15", "Economy_15", "Family_15", "Life_Expectancy_15", "Freedom_15", "Trust_15", "Generosity_15" , "Dystopia_15")

dataframe_2016.columns = ("Country", "Region", "Happiness_Rank_16", "Happiness_Score_16",  "Economy_16", "Family_16", "Life_Expectancy_16", "Freedom_16", "Trust_16", "Generosity_16", "Dystopia_16")

dataframe_2017.columns = ("Country", "Region", "Happiness_Rank_17", "Happiness_Score_17",  "Economy_17", "Family_17", "Life_Expectancy_17", "Freedom_17", "Generosity_17", "Trust_17", "Dystopia_17")
```

```python
# Plot the happiness ranking on a world map for 2017

data_for_plot = dict(type = 'choropleth', 
           locations = dataframe_2017['Country'],
           locationmode = 'country names',
           z = dataframe_2017['Happiness_Rank_17'], 
           text = dataframe_2017['Country'],
          colorscale = 'YlGnBu', reversescale = False)
mapspec = dict(title = 'World Happiness Ranking 2017', 
             geo = dict(showframe = False, projection = {'type': 'natural earth'}))

HappyMap = go.Figure(data = [data_for_plot], layout=mapspec)
iplot(HappyMap)
```

```python
#Density of happiness scores for year 2017

sns.distplot(dataframe_2017['Happiness_Score_17'], norm_hist= False)
# The histogram height shows density fraction
```

```python
# Data Frame with Happiness Scores for each year:

happiness_score_all_df = pd.DataFrame({'Happiness Score 2015': dataframe_2015['Happiness_Score_15'], 'Happiness Score 2016': dataframe_2016['Happiness_Score_16'], 'Happiness Score 2017': dataframe_2017['Happiness_Score_17']})
```

```python
"""
Combining data for all three years
First merge data for 2015 and 2016, later merge it with 2017 data
"""

#merge data for 2015 and 2016
dataframe_1516 = pd.merge(dataframe_2015, dataframe_2016, on = ['Country'])

#merge data for 2015 and 2017
dataframe_1517 = pd.merge(dataframe_2015, dataframe_2017, on = ['Country'])

#Combining data for all three years
dataframe_151617 = pd.merge(dataframe_1516, dataframe_2017, on = ['Country'])

dataframe_151617.to_csv("Country_Check_Data.csv")
```

```python
#dataframe for storing difference in happiness score 
dataframe_diff = pd.DataFrame()
```

```python
dataframe_diff['Country'] = dataframe_1517['Country']

dataframe_diff['Happiness_Rank_diff'] = dataframe_1517['Happiness_Rank_17'] - dataframe_1517['Happiness_Rank_15']

dataframe_diff['Happiness_Score_diff'] = dataframe_1517['Happiness_Score_17'] - dataframe_1517['Happiness_Score_15']

dataframe_diff['Economy_diff'] = dataframe_1517['Economy_17'] - dataframe_1517['Economy_15']

dataframe_diff['Family_diff'] = dataframe_1517['Family_17'] - dataframe_1517['Family_15']

dataframe_diff['Life_Expectancy_diff'] = dataframe_1517['Life_Expectancy_17'] - dataframe_1517['Life_Expectancy_15']

dataframe_diff['Freedom_diff'] = dataframe_1517['Freedom_17'] - dataframe_1517['Freedom_15']

dataframe_diff['Trust_diff'] = dataframe_1517['Trust_17'] - dataframe_1517['Trust_15']

dataframe_diff['Generosity_diff'] = dataframe_1517['Generosity_17'] - dataframe_1517['Generosity_15']
```

```python

#Sorting the df_diff file by the difference in happiness scores

dataframe_diff_sorted = dataframe_diff.sort_values('Happiness_Score_diff')
```

```python
#Boxplot comparing distribution of Happiness Score for the three years

ax = plt.axes()
sns.boxplot(happiness_score_all_df)
ax.set_title('Happiness Scores for 2015, 2016 & 2017', fontsize = 15 ,fontweight = 'bold')
plt.show()
```

```python
#Boxplot for 2017

ax = plt.axes()
sns.boxplot(dataframe_2017[['Economy_17', 'Family_17', 'Life_Expectancy_17', 'Freedom_17', 'Trust_17', 'Generosity_17']])
ax.set_title('Factors contributing to Happiness Scores (2017)', fontsize = 15,fontweight = 'bold')
ax.figure.set_size_inches(10, 7)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.show()
```

```python
#divide data into group by the amount of decrease or increase they saw in year 2015 and 2017.

df_increase = dataframe_diff_sorted.loc[dataframe_diff_sorted['Happiness_Score_diff'] >= 0.4]
```

```python
#Scatter plot of countries with increase in happiness score greater than 0.4
q = sns.stripplot(x="Country", y="Happiness_Score_diff", data=df_increase, jitter=True)
plt.xticks(rotation=90)
plt.title('Increase above or equal to 0.4',fontsize=20,fontweight = 'bold')
plt.show()
```

```python
#Getting the Top N countries with the highest decrease in happiness scores from 2015 to 2017

N = 10
Top10_Decrease = dataframe_diff_sorted[0:N]
Top10_Increase = dataframe_diff_sorted[dataframe_diff_sorted.shape[0]-N:]

#Indexing the Top10_Decrease and Top10_Increase datasets by country, so as to plot in a heatmap later

Top10_Decrease.index = Top10_Decrease['Country']

del Top10_Decrease['Country']

Top10_Increase.index = Top10_Increase['Country']

del Top10_Increase['Country']
```

```python
#Plotting Top 10 countries with highest decrease in happiness against the difference in explanatory factors for happiness scores

ax = plt.axes()
sns.heatmap(Top10_Decrease[Top10_Decrease.columns[2:]], annot = True, cmap = 'magma', ax = ax)
ax.set_title("\n".join(wrap('Top 10 Countries With Highest Decrease in Happiness Scores')), fontsize = 15 ,fontweight = 'bold')
ax.figure.set_size_inches(10, 7)
plt.xticks(size = 12, rotation = 45)
plt.yticks(size = 12)
plt.show()
```

```python
#Plotting Top 10 countries with highest increase in happiness against the difference in explanatory factors for happiness scores

ax = plt.axes()
sns.heatmap(Top10_Increase[Top10_Increase.columns[2:]], annot = True, cmap = 'magma', ax = ax)
ax.set_title("\n".join(wrap('Top 10 Countries With Highest Increase in Happiness Scores'))
             , fontsize = 15, fontweight = 'bold')
ax.figure.set_size_inches(10, 7)
plt.xticks(size = 12, rotation = 45)
plt.yticks(size = 12)
plt.show()
```

```python
#changing column name
dataframe_2015 = dataframe_2015.rename(columns={'Happiness_Score_15': 'Happiness_Score', 'Economy_15': 'Economy'})
dataframe_2016 = dataframe_2016.rename(columns={'Happiness_Score_16': 'Happiness_Score', 'Economy_16': 'Economy'})
dataframe_2017 = dataframe_2017.rename(columns={'Happiness_Score_17': 'Happiness_Score', 'Economy_17': 'Economy'})
```

```python
dataframe_2017 = dataframe_2017.loc[dataframe_2017['Country'] == 'India']
dataframe_2016 = dataframe_2016.loc[dataframe_2016['Country'] == 'India']
dataframe_2015 = dataframe_2015.loc[dataframe_2015['Country'] == 'India']

dataframe_2017 = dataframe_2017.iloc[:, 4:10]
dataframe_2016 = dataframe_2016.iloc[:, 4:10]
dataframe_2015 = dataframe_2015.iloc[:, 4:10]


columns = tuple(dataframe_2016.dtypes.index)
columns1 = tuple(dataframe_2017.dtypes.index)

dataframe_2015 = dataframe_2015.iloc[:, :].values
dataframe_2016 = dataframe_2016.iloc[:, :].values
dataframe_2017 = dataframe_2017.iloc[:, :].values
```

```python
N = dataframe_2015[0, :].shape[0]
N1 = dataframe_2016[0, :].shape[0]
N3 = dataframe_2017[0, :].shape[0]


rcParams['figure.figsize'] = 20, 10
plt.rc('font', weight='bold')
plt.rc('font', weight='bold')

ind = np.arange(N)
ind1=np.arange(N1)
ind2=np.arange(N3)
width = 0.3
fig, ax = plt.subplots()

bar1 = ax.bar(ind, dataframe_2015[0, :].tolist(), width, color = 'yellow')
bar2 = ax.bar(ind1 + width, dataframe_2016[0, :].tolist(), width, color = 'darkgreen')
bar3 = ax.bar(ind1 + width*2, dataframe_2017[0, :].tolist(), width, color = 'darkblue')

ax.set_ylabel('Scores',fontsize=16,fontweight='bold')
ax.set_title('Factors affecting Happiness Score in India',fontsize=20,fontweight='bold')
ax.set_xticks(ind + width*2 / 2)
ax.set_xticklabels(columns,fontsize=16)

ax.legend((bar1[0], bar2[0],bar3[0]), ('2015', '2016','2017'), loc='upper center')

plt.show()
```

```python
#Computing average of all factors by region

df_15_avg = dataframe_2015.groupby(['Region']).mean()
df_16_avg = dataframe_2016.groupby(['Region']).mean()
df_17_avg = dataframe_2017.groupby(['Region']).mean()


#Adding a column for all averaged data sets to represent year

df_15_avg['Year'] = '2015'
df_16_avg['Year'] = '2016'
df_17_avg['Year'] = '2017'

#Combining all averaged data sets

df_avg = pd.concat([df_15_avg, df_16_avg, df_17_avg])
```

```python
#Average happiness scores by region for all years

ax = plt.axes()
sns.barplot(x=df_avg.index, y="Happiness_Score", hue = 'Year',palette='bright', data=df_avg)
ax.set_title('Average Happiness Scores by Region for all Years',  fontsize = 15, fontweight = 'bold')
ax.figure.set_size_inches(10, 6)
plt.xticks(rotation = 45)
plt.yticks(size = 12)
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
plt.show()
```

The barchart above represents categorical data(regions in this case) and the average value of the factors affecting the happiness score across three years. You can clearly see the increase or decrease across the years. The Sub-Saharan African Region has the lowest average Happiness Score followed by Southern Asia.
